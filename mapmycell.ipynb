```python
import scanpy as sc
import anndata
import os

# helper to read a 10x “filtered_feature_bc_matrix” directory
def read_10x(dir_path, var_names="gene_symbols"):
    return sc.read_10x_mtx(
        os.path.join(dir_path, "filtered_feature_bc_matrix"),
        var_names=var_names,
        cache=True
    )

```

```python
# --- Channel 1 ---
adata1 = read_10x("/gpfs/group/jin/boli/250710_GEMX_multiBrainReg/250710/250710_Ch1_R2_12/outs")
# load the CRISPR counts into a second layer
adata1.layers["CRISPR"] = read_10x(
    "/gpfs/group/jin/boli/250710_GEMX_multiBrainReg/250710/250710_Ch1_R2_12/outs",
    var_names="gene_symbols"
).X  # just change the var_names if needed to match the guide-capture features
adata1.obs["sample"] = "Ch1"
adata1.write("Channel1.h5ad")

```

```python
# --- Channel 2 ---
adata2 = read_10x("/gpfs/group/jin/boli/250710_GEMX_multiBrainReg/250710/250710_Ch2_R2_12/outs")
adata2.layers["CRISPR"] = read_10x(
   "/gpfs/group/jin/boli/250710_GEMX_multiBrainReg/250710/250710_Ch2_R2_12/outs",
).X
adata2.obs["sample"] = "Ch2"
adata2.write("Channel2.h5ad")

# --- Channel 3 ---
adata3 = read_10x("/gpfs/group/jin/boli/250710_GEMX_multiBrainReg/250710/250710_Ch3_R2_12/outs")
adata3.layers["CRISPR"] = read_10x(
    "/gpfs/group/jin/boli/250710_GEMX_multiBrainReg/250710/250710_Ch3_R2_12/outs",
).X
adata3.obs["sample"] = "Ch3"
adata3.write("Channel3.h5ad")
```

```python
import scanpy as sc

adata1 = sc.read_h5ad("Channel1.h5ad")
adata2 = sc.read_h5ad("Channel2.h5ad")
adata3 = sc.read_h5ad("Channel3.h5ad")


adata_all = sc.concat(
    [adata1, adata2, adata3],
    join="outer",
    label="sample",
    keys=["Ch1", "Ch2", "Ch3"]
)

print(adata_all)  

adata_all.write("all_channels.h5ad", compression="gzip")
```
